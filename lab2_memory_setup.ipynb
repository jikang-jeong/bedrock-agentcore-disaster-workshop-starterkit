```python
from dotenv import load_dotenv
import os
import boto3

load_dotenv()

REGION = os.getenv('AWS_REGION', 'us-west-2')
print(f"ğŸŒ Region: {REGION}")
```

## 2-2. Memory ìƒì„±

```python
from bedrock_agentcore_starter_toolkit.operations.memory.manager import MemoryManager
from bedrock_agentcore_starter_toolkit.operations.memory.models.strategies import SemanticStrategy

print("ğŸ§  Memory ë¦¬ì†ŒìŠ¤ ìƒì„± ì¤‘...")

memory_manager = MemoryManager(region_name=REGION)

memory = memory_manager.get_or_create_memory(
    name="customMemory",
    description="í™”ì¬ ëŒ€ì‘ ì–´ì‹œìŠ¤í„´íŠ¸ ë©”ëª¨ë¦¬ ì €ì¥ì†Œ",
    strategies=[
        SemanticStrategy(
            name="semanticLongTermMemory",
            namespaces=['/actors/{actorId}'],
        )
    ],
    event_expiry_days=30
)

MEMORY_ID = memory.get('id')
print(f"\nâœ… Memory ìƒì„± ì™„ë£Œ!")
print(f"ğŸ“ Memory ID: {MEMORY_ID}")
```

## 2-3. Memory IDë¥¼ .envì— ì €ì¥

```python
# .env íŒŒì¼ ì½ê¸°
with open('.env', 'r') as f:
    lines = f.readlines()

# MEMORY_ID ì¶”ê°€ ë˜ëŠ” ì—…ë°ì´íŠ¸
found = False
for i, line in enumerate(lines):
    if line.startswith('MEMORY_ID='):
        lines[i] = f"MEMORY_ID={MEMORY_ID}\n"
        found = True
        break

if not found:
    lines.append(f"MEMORY_ID={MEMORY_ID}\n")

# .env íŒŒì¼ ì €ì¥
with open('.env', 'w') as f:
    f.writelines(lines)

print(f"âœ… Memory IDë¥¼ .envì— ì €ì¥ ì™„ë£Œ")
print(f"ğŸ“ Memory ID: {MEMORY_ID}")
```
